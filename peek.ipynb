```python
from transformers import ChameleonProcessor, ChameleonForConditionalGeneration

local_path = "/aifs4su/yaodong/projects/hantao/anole/facilitating_image_generation/model/chameleon_hf_0830_4k"

processor = ChameleonProcessor.from_pretrained(local_path)
model = ChameleonForConditionalGeneration.from_pretrained(local_path, device_map="auto")
```

```python
model.config
```

```python
state_dict = model.state_dict()

for key in state_dict:
    print(key)
    print(state_dict[key].shape)
    print("="*50)
```

```python
img_token_list = model.model.vocabulary_mapping.image_tokens

# save to file
with open("img_token_list.txt", "w") as f:
    f.write(str(img_token_list))
```

```python
prompt = "Where is the capital of Germany?"
input = processor(prompt, return_tensors="pt").to(
            model.device, dtype=model.dtype
        )
input
```

```python
output = model.generate(**input, max_new_tokens=10, multimodal_generation_mode="text-only", temperature=0.05, do_sample=True)
print(processor.tokenizer.decode(output[0], skip_special_tokens=True))
```
